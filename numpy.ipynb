N-dimensional array ndarray from numpy is a library that allows to work with array of n dimension. Arrays with n>2 are called tensor.

**Declaration of n-dimensional arrays in numpy**

```python
import numpy as np

a = np.array([[1,2,3],
              [4,5,5]])

print(a.shape, a.dtype)

n_rows, n_colons = 4,5
a = np.zeros(shape = (n_rows, n_colons), dtype=int)#the shape is a tuple
print(a)

a = np.ones(shape = (n_rows, n_colons), dtype=int)#the shape is a tuple
print(a)

a = np.eye(n_rows, n_colons, dtype=int)#identity matrix, does not want the shape argument
print(a)

################################Random###########################

a = np.random.randn(n_rows, n_colons)#random numbers from Normal distribution with zero mean and unit variance

a = np.random.rand(n_rows, n_colons)#random numbers form Unifom distribution in [0,1]

a = np.random.randint(0, 5, [n_rows, n_colons])#invervall of integers to generate and shape as a list
```

**Array Indexing**

```python
a[0,0]#row-column, nomal indexing
print(a[0, :])#the column requires the full row (all columns in row 0)
a[:,0]#the column requires the full column (all rows in columsn 0)

a[0:2, 0:2]#we can extract a submatrix withc the slicing operator

b = np.array([1,1,0])
a[b==1, :]# this takes the first two rows and every column, this notation is going to extract the rows for which the correspondet value in b satisfies the condition. What happens is that we are inputing as the firs element a boolean vector
          # the boolean vector i generated by the operation b==1 wicht return a vector containg the value true for every element that satisfais the condition. the extraction of a submatrix works with the true values
```

```python
from matplotlib import pyplot as plt
from scipy.stats import norm

x = np.random.randn(1000,1)
#print(x)

plt.plot(x, '.')

#compute the distribution of the values
plt.figure()#new image
plt.hist(x)

#pdf 
pdf_x = norm.pdf(x)#return the probability value for the correspondent x

plt.figure()
plt.plot(x, pdf_x, '.')
```

```python
x = np.random.random(10000)
print(x.shape)

plt.plot(x, '.')

#compute the distribution of the values
plt.figure()#new image
plt.hist(x)

```

**CDF from data**

```python
x = x.ravel()#squash all the element in a single row
th = 0

#how do I find the elements that are below the given threshold?
print(sum(x<0.5))
#cdf for X=x
value=0.5

print(sum(x<value)/x.size)

#CDF

th_values = np.linspace(0,1,1000,endpoint = True)#x values for the cdf function
cdf_prob = np.zeros(shape = th_values.shape)
#print(th_values.shape)

for i, th in enumerate(th_values):
  cdf_prob[i] = sum(x<th)/x.size

plt.figure()
plt.plot(th_values, cdf_prob)
```
